```python
!pip install qiskit qiskit_aer qiskit-ibm-runtime matplotlib
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile
from qiskit_aer import AerSimulator
from qiskit_aer.noise import NoiseModel, depolarizing_error, amplitude_damping_error
from qiskit_ibm_runtime import QiskitRuntimeService, Sampler
from qiskit.quantum_info import Operator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
import numpy as np

def oracle_constant(qc, ancilla, value=0):
    """Constant oracle: f(x)=0 or f(x)=1"""
    if value == 1:
        qc.x(ancilla)

def oracle_balanced_parity(qc, inputs, ancilla):
    """Balanced oracle: f(x) = x0 XOR x1 XOR ... XOR xn"""
    for q in inputs:
        qc.cx(q, ancilla)

def deutsch_jozsa_circuit(n, oracle_func, *oracle_args):
    """Build Deutsch-Jozsa circuit with given oracle"""
    qreg = QuantumRegister(n + 1, "q")
    creg = ClassicalRegister(n, "c")
    qc = QuantumCircuit(qreg, creg)

    inputs = list(range(n))
    ancilla = n

    # Initialize
    qc.x(ancilla)
    qc.h(qreg)

    # Apply oracle
    oracle_func(qc, *oracle_args)

    # Hadamard on inputs
    for q in inputs:
        qc.h(q)

    # Measure
    qc.measure(inputs, creg)
    return qc


def run_dj_basic(qc):
    """Basic execution with result interpretation"""
    simulator = AerSimulator()
    tqc = transpile(qc, simulator)
    job = simulator.run(tqc, shots=1024)
    result = job.result()
    counts = result.get_counts()

    print("Measurement counts:", counts)
    plot_histogram(counts)
    plt.show()

    n = qc.num_clbits
    if counts.get("0" * n, 0) == 1024:
        print("âœ… Function is CONSTANT")
    else:
        print("âœ… Function is BALANCED")

    return counts


# TASK 1: MODIFY THE ORACLE - CUSTOM BALANCED ORACLE

print("\n" + "="*80)
print("TASK 1: CUSTOM BALANCED ORACLE")
print("="*80 + "\n")

def oracle_balanced_custom(qc, inputs, ancilla):
    n = len(inputs)

    if n == 2:
        # For 2 qubits: flip for |01âŸ© and |10âŸ©
        qc.x(inputs[0])
        qc.ccx(inputs[0], inputs[1], ancilla)
        qc.x(inputs[0])
        qc.x(inputs[1])
        qc.ccx(inputs[0], inputs[1], ancilla)
        qc.x(inputs[1])

    elif n == 3:
        # For 3 qubits: flip for specific patterns
        # Pattern 1: |001âŸ©
        qc.x(inputs[0])
        qc.x(inputs[1])
        qc.ccx(inputs[0], inputs[1], ancilla)
        qc.cx(inputs[2], ancilla)
        qc.ccx(inputs[0], inputs[1], ancilla)
        qc.x(inputs[1])
        qc.x(inputs[0])

        # Pattern 2: |010âŸ©
        qc.x(inputs[0])
        qc.x(inputs[2])
        qc.ccx(inputs[0], inputs[2], ancilla)
        qc.cx(inputs[1], ancilla)
        qc.ccx(inputs[0], inputs[2], ancilla)
        qc.x(inputs[2])
        qc.x(inputs[0])

        # Pattern 3: |100âŸ©
        qc.x(inputs[1])
        qc.x(inputs[2])
        qc.ccx(inputs[1], inputs[2], ancilla)
        qc.cx(inputs[0], ancilla)
        qc.ccx(inputs[1], inputs[2], ancilla)
        qc.x(inputs[2])
        qc.x(inputs[1])

        # Pattern 4: |111âŸ©
        qc.ccx(inputs[0], inputs[1], ancilla)
        qc.cx(inputs[2], ancilla)
        qc.ccx(inputs[0], inputs[1], ancilla)

    else:
        # Generic approach for other n: alternate based on MSB and LSB
        qc.cx(inputs[0], ancilla)
        if n > 1:
            qc.ccx(inputs[0], inputs[-1], ancilla)

# Run Task 1
print("Testing custom balanced oracle with n=3 qubits:\n")
n = 3
qc_task1 = deutsch_jozsa_circuit(n, oracle_balanced_custom, list(range(n)), n)
print("Circuit Diagram:")
print(qc_task1.draw(fold=-1))
print("\nExecution Results:")
run_dj_basic(qc_task1)

# TASK 2: CHANGE THE NUMBER OF INPUT QUBITS

print("\n\n" + "="*80)
print("TASK 2: VARYING NUMBER OF INPUT QUBITS")
print("="*80 + "\n")

def analyze_circuit_properties(qc, n):
    """Analyze and print circuit properties"""
    print(f"\n--- Circuit Analysis for n={n} ---")
    print(f"Number of qubits: {qc.num_qubits}")
    print(f"Number of classical bits: {qc.num_clbits}")
    print(f"Circuit depth: {qc.depth()}")
    print(f"Total gate count: {sum(qc.count_ops().values())}")
    print(f"Gate breakdown: {qc.count_ops()}")

# Run Task 2
test_values = [2, 3, 4, 5]

for n in test_values:
    print(f"\n{'='*60}")
    print(f"Testing with n = {n} input qubits")
    print(f"{'='*60}")

    # Test with balanced oracle
    qc = deutsch_jozsa_circuit(n, oracle_balanced_parity, list(range(n)), n)

    # Analyze circuit properties
    analyze_circuit_properties(qc, n)

    # Draw circuit (for smaller n)
    if n <= 3:
        print("\nCircuit diagram:")
        print(qc.draw(fold=-1))

    # Execute and observe results
    print(f"\nExecuting Deutsch-Jozsa with n={n}:")
    run_dj_basic(qc)

    print(f"\nğŸ“Š Observation: For balanced parity oracle with {n} qubits,")
    print(f"expected output is a non-zero bitstring pattern.")



# TASK 3: ADD NOISE SIMULATION

print("\n\n" + "="*80)
print("TASK 3: NOISE SIMULATION")
print("="*80 + "\n")

def build_noise_model(noise_level='medium'):
    """
    Create a custom noise model with depolarizing and damping errors
    noise_level: 'low', 'medium', or 'high'
    """
    noise_params = {
        'low': {'depol_1q': 0.001, 'depol_2q': 0.01, 'damping': 0.01},
        'medium': {'depol_1q': 0.01, 'depol_2q': 0.05, 'damping': 0.03},
        'high': {'depol_1q': 0.05, 'depol_2q': 0.15, 'damping': 0.1}
    }

    params = noise_params[noise_level]
    noise_model = NoiseModel()

    # Depolarizing error for single-qubit gates
    error_1q = depolarizing_error(params['depol_1q'], 1)
    noise_model.add_all_qubit_quantum_error(error_1q, ['h', 'x'])

    # Depolarizing error for two-qubit gates
    error_2q = depolarizing_error(params['depol_2q'], 2)
    noise_model.add_all_qubit_quantum_error(error_2q, ['cx'])

    # Amplitude damping for measurements
    error_meas = amplitude_damping_error(params['damping'])
    noise_model.add_all_qubit_quantum_error(error_meas, 'measure')

    return noise_model

def run_dj_noisy(qc, noise_model=None, noise_label=""):
    """Run circuit with optional noise model"""
    simulator = AerSimulator(noise_model=noise_model)
    tqc = transpile(qc, simulator)
    job = simulator.run(tqc, shots=1024)
    result = job.result()
    counts = result.get_counts()

    print(f"\n{noise_label}")
    print("Measurement counts:", counts)
    plot_histogram(counts)
    plt.title(noise_label)
    plt.show()

    n = qc.num_clbits
    zero_count = counts.get("0" * n, 0)
    print(f"Count of all-zero measurement: {zero_count}/1024 ({100*zero_count/1024:.1f}%)")

    if zero_count > 900:
        print("âœ… Function appears CONSTANT (high confidence)")
    elif zero_count < 100:
        print("âœ… Function appears BALANCED (high confidence)")
    else:
        print("âš ï¸  Result inconclusive due to noise")

    return counts

# Run Task 3
n = 3

print("\nTesting CONSTANT Oracle with Different Noise Levels:")
print("="*60)
qc_const = deutsch_jozsa_circuit(n, oracle_constant, n, 0)

run_dj_noisy(qc_const, noise_model=None, noise_label="Noiseless Simulation")
noise_low = build_noise_model('low')
run_dj_noisy(qc_const, noise_model=noise_low, noise_label="Low Noise Simulation")
noise_med = build_noise_model('medium')
run_dj_noisy(qc_const, noise_model=noise_med, noise_label="Medium Noise Simulation")
noise_high = build_noise_model('high')
run_dj_noisy(qc_const, noise_model=noise_high, noise_label="High Noise Simulation")

print("\n\nTesting BALANCED Oracle with Different Noise Levels:")
print("="*60)
qc_bal = deutsch_jozsa_circuit(n, oracle_balanced_parity, list(range(n)), n)

run_dj_noisy(qc_bal, noise_model=None, noise_label="Noiseless Simulation")
run_dj_noisy(qc_bal, noise_model=noise_med, noise_label="Medium Noise Simulation")
run_dj_noisy(qc_bal, noise_model=noise_high, noise_label="High Noise Simulation")


# TASK 4: RUN ON IBM QUANTUM DEVICE
print("\n\n" + "="*80)
print("TASK 4: IBM QUANTUM HARDWARE EXECUTION")
print("="*80 + "\n")

def run_on_simulator(qc):
    """Run on local simulator"""
    print("\n--- Running on AerSimulator ---")
    simulator = AerSimulator()
    tqc = transpile(qc, simulator)
    job = simulator.run(tqc, shots=1024)
    result = job.result()
    counts = result.get_counts()

    print("Simulator counts:", counts)
    plot_histogram(counts, title="Simulator Results")
    plt.show()
    return counts

def run_on_ibm_device(qc, backend_name=None):
    """Run on IBM Quantum device"""
    try:
        service = QiskitRuntimeService(channel="ibm_quantum")

        if backend_name is None:
            backend = service.least_busy(
                simulator=False,
                operational=True,
                min_num_qubits=qc.num_qubits
            )
        else:
            backend = service.backend(backend_name)

        print(f"\n--- Running on IBM Quantum Device: {backend.name} ---")
        print(f"Backend properties: {backend.num_qubits} qubits")

        tqc = transpile(qc, backend, optimization_level=3)
        print(f"Transpiled circuit depth: {tqc.depth()}")

        sampler = Sampler(backend)
        job = sampler.run(tqc, shots=1024)

        print(f"Job ID: {job.job_id()}")
        print("Waiting for results...")

        result = job.result()
        counts = result.quasi_dists[0].binary_probabilities()
        counts_dict = {k: int(v * 1024) for k, v in counts.items()}

        print("IBM Quantum counts:", counts_dict)
        plot_histogram(counts_dict, title=f"IBM Quantum Results ({backend.name})")
        plt.show()

        return counts_dict

    except Exception as e:
        print(f"Error running on IBM Quantum: {e}")
        print("Make sure you have:")
        print("1. Saved your IBM Quantum token")
        print("2. Have access to IBM Quantum devices")
        print("3. Internet connection is active")
        return None

def compare_results(sim_counts, ibm_counts, qc):
    """Compare simulator and real hardware results"""
    if ibm_counts is None:
        print("\nCannot compare - IBM Quantum execution failed")
        return

    print("\n" + "="*60)
    print("COMPARISON: Simulator vs IBM Quantum Hardware")
    print("="*60)

    n = qc.num_clbits
    all_zero = "0" * n

    sim_zero_pct = 100 * sim_counts.get(all_zero, 0) / 1024
    ibm_zero_pct = 100 * ibm_counts.get(all_zero, 0) / 1024

    print(f"\nAll-zero measurement percentage:")
    print(f"  Simulator: {sim_zero_pct:.1f}%")
    print(f"  IBM Quantum: {ibm_zero_pct:.1f}%")
    print(f"  Difference: {abs(sim_zero_pct - ibm_zero_pct):.1f}%")

# Run Task 4
n = 3

print("\nTesting CONSTANT Oracle:")
qc_const = deutsch_jozsa_circuit(n, oracle_constant, n, 0)
print(qc_const.draw(fold=-1))

sim_counts_const = run_on_simulator(qc_const)

print("\n" + "="*60)
print("\nTesting BALANCED Oracle:")
qc_bal = deutsch_jozsa_circuit(n, oracle_balanced_parity, list(range(n)), n)
print(qc_bal.draw(fold=-1))

sim_counts_bal = run_on_simulator(qc_bal)

print("\nğŸ“ NOTE: To run on real IBM Quantum hardware:")
print("1. Uncomment the QiskitRuntimeService.save_account() line")
print("2. Add your IBM Quantum token from https://quantum.ibm.com/")
print("3. Uncomment the run_on_ibm_device() calls")

# TASK 5: CIRCUIT ANALYSIS - UNITARY REPRESENTATION
print("\n\n" + "="*80)
print("TASK 5: CIRCUIT ANALYSIS - UNITARY REPRESENTATION")
print("="*80 + "\n")

def analyze_oracle_unitary(oracle_func, n, *oracle_args):
    """Extract and analyze the unitary matrix of an oracle"""
    print(f"\n{'='*70}")
    print(f"Analyzing Oracle: {oracle_func.__name__}")
    print(f"{'='*70}")

    qreg = QuantumRegister(n + 1, "q")
    qc = QuantumCircuit(qreg)
    oracle_func(qc, *oracle_args)

    print(f"\nOracle Circuit (n={n} input qubits):")
    print(qc.draw(fold=-1))

    print(f"\nGate Statistics:")
    print(f"  Total gates: {sum(qc.count_ops().values())}")
    print(f"  Gate breakdown: {qc.count_ops()}")
    print(f"  Circuit depth: {qc.depth()}")

    try:
        operator = Operator(qc)
        unitary = operator.data

        print(f"\nUnitary Matrix Properties:")
        print(f"  Matrix shape: {unitary.shape}")
        print(f"  Matrix is unitary: {operator.is_unitary()}")

        identity_check = np.allclose(
            unitary @ unitary.conj().T,
            np.eye(len(unitary))
        )
        print(f"\nQuantum Properties:")
        print(f"  U â€  U = I (identity): {identity_check}")

        print(f"\nUnitary Matrix (first 8x8 block):")
        print(f"{np.array2string(unitary[:8, :8], precision=3, suppress_small=True)}")

        analyze_oracle_behavior(unitary, n)

        return unitary

    except Exception as e:
        print(f"Error computing unitary: {e}")
        return None

def analyze_oracle_behavior(unitary, n):
    """Analyze how the oracle transforms basis states"""
    print(f"\nOracle Function Behavior f(x):")
    print(f"{'Input (x)':>12} | {'Ancilla Flip?':>15} | {'f(x)'}")
    print("-" * 45)

    num_inputs = 2 ** n
    flips = []

    for i in range(num_inputs):
        state_00 = i * 2
        state_01 = i * 2 + 1

        flip_00_to_01 = abs(unitary[state_01, state_00]) > 0.5
        flip_01_to_00 = abs(unitary[state_00, state_01]) > 0.5

        if flip_00_to_01 and not flip_01_to_00:
            f_x = 1
            flips.append(1)
        else:
            f_x = 0
            flips.append(0)

        binary_x = format(i, f'0{n}b')
        flip_status = "YES" if f_x == 1 else "NO"
        print(f"{binary_x:>12} | {flip_status:>15} | {f_x}")

    num_ones = sum(flips)
    total = len(flips)

    print(f"\n{'='*45}")
    if num_ones == 0 or num_ones == total:
        print(f"âœ… Oracle is CONSTANT: f(x) = {flips[0]} for all x")
    elif num_ones == total // 2:
        print(f"âœ… Oracle is BALANCED: {num_ones}/{total} inputs produce f(x)=1")
    else:
        print(f"âš ï¸  Oracle is neither constant nor balanced")
    print(f"{'='*45}")

def explain_oracle_implementation(oracle_func, n):
    """Provide explanation of how the oracle implements f(x)"""
    print(f"\n{'='*70}")
    print(f"Implementation Explanation: {oracle_func.__name__}")
    print(f"{'='*70}\n")

    if oracle_func.__name__ == 'oracle_constant':
        print("This oracle implements f(x) = constant.")
        print("Implementation: Uses a single X gate (or no gate)")
        print("  - If value=0: No operation â†’ f(x)=0 for all x")
        print("  - If value=1: X gate on ancilla â†’ f(x)=1 for all x")
        print("\nMechanism: The ancilla qubit state is set independently")
        print("of the input qubits, resulting in constant output.")

    elif oracle_func.__name__ == 'oracle_balanced_parity':
        print("This oracle implements f(x) = xâ‚€ âŠ• xâ‚ âŠ• ... âŠ• xâ‚™â‚‹â‚ (XOR parity).")
        print(f"Implementation: Chain of {n} CX gates")
        print("  - Each CX(xáµ¢, ancilla) flips ancilla if xáµ¢=1")
        print(f"  - Total flips = number of 1s in input")
        print("  - f(x)=1 if odd number of 1s, f(x)=0 if even")
        print("\nMechanism: This creates a balanced function because")
        print(f"exactly half ({2**(n-1)} out of {2**n}) of all inputs")
        print("have odd parity (odd number of 1s).")

    elif oracle_func.__name__ == 'oracle_balanced_custom':
        print("This oracle implements a custom balanced function.")
        print("Implementation: Uses CCX (Toffoli) and CX gates")
        print("  - CCX(xâ‚€, xâ‚, ancilla): Flips ancilla if xâ‚€ AND xâ‚ = 1")
        if n == 3:
            print("  - CX(xâ‚‚, ancilla): Additionally flips if xâ‚‚ = 1")
            print("\nFunction: f(x) = (xâ‚€ âˆ§ xâ‚) âŠ• xâ‚‚")
        print("\nMechanism: Controlled operations create specific")
        print("patterns that result in exactly half the inputs")
        print("producing f(x)=1.")

def visualize_unitary_matrix(unitary, title):
    """Create heatmap visualization of unitary matrix"""
    if unitary is None:
        return

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    im1 = ax1.imshow(np.abs(unitary), cmap='viridis', aspect='auto')
    ax1.set_title(f'{title}\nMagnitude |U|')
    ax1.set_xlabel('Column')
    ax1.set_ylabel('Row')
    plt.colorbar(im1, ax=ax1)

    phase = np.angle(unitary)
    im2 = ax2.imshow(phase, cmap='twilight', aspect='auto', vmin=-np.pi, vmax=np.pi)
    ax2.set_title(f'{title}\nPhase âˆ U')
    ax2.set_xlabel('Column')
    ax2.set_ylabel('Row')
    plt.colorbar(im2, ax=ax2)

    plt.tight_layout()
    plt.show()

# Run Task 5
n = 3

# Analyze Constant Oracle (f(x) = 0)
unitary_const_0 = analyze_oracle_unitary(oracle_constant, n, n, 0)
explain_oracle_implementation(oracle_constant, n)

# Analyze Constant Oracle (f(x) = 1)
print("\n\n")
unitary_const_1 = analyze_oracle_unitary(oracle_constant, n, n, 1)

# Analyze Balanced Parity Oracle
print("\n\n")
unitary_bal_parity = analyze_oracle_unitary(
    oracle_balanced_parity, n, list(range(n)), n
)
explain_oracle_implementation(oracle_balanced_parity, n)

# Analyze Custom Balanced Oracle
print("\n\n")
unitary_bal_custom = analyze_oracle_unitary(
    oracle_balanced_custom, n, list(range(n)), n
)
explain_oracle_implementation(oracle_balanced_custom, n)

# Visualize unitaries
if unitary_const_0 is not None:
    visualize_unitary_matrix(unitary_const_0, "Constant Oracle (f=0)")

if unitary_bal_parity is not None:
    visualize_unitary_matrix(unitary_bal_parity, "Balanced Parity Oracle")

if unitary_bal_custom is not None:
    visualize_unitary_matrix(unitary_bal_custom, "Custom Balanced Oracle")

print("\n\n" + "="*80)
print("ALL 5 TASKS COMPLETED!")
print("="*80)
print("\nâœ… Task 1: Custom balanced oracle implemented and tested")
print("âœ… Task 2: Circuit tested with n=2,3,4,5 qubits")
print("âœ… Task 3: Noise simulation with low/medium/high noise levels")
print("âœ… Task 4: IBM Quantum framework ready (add token to run)")
print("âœ… Task 5: Circuit analysis with unitary matrices and explanations")
print("\n" + "="*80)
```
